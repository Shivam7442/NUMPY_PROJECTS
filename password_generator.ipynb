## Making datasets for lower upper special and digit

```python
import string
import numpy as np
import random
```

```python
upper="".join(np.array(list(string.ascii_uppercase)))
print(upper)
```

```python
lower="".join(np.array(list(string.ascii_lowercase)))
digits="".join(np.array(list(string.digits)))
special="".join(np.array(list(string.punctuation )))
special.replace(" ","")
```

```python
n=int(input("Length of the password(4,8,12,16):"))
m=int(input("No of passwords to be generated:"))
```

```python
r=int(n/4)
passwords=[]

for _ in range(m):
    
    password=[]
    password.extend(random.sample(upper,r))
    password.extend(random.sample(lower,r))
    password.extend(random.sample(digits,r))
    password.extend(random.sample(special,r))
    random.shuffle(password)
    passwords.append("".join(password))

print(passwords)
```

# Code generated by AI ,it uses NUMPY extensively 

```python
#import numpy as np
#import string

# Character sets as NumPy arrays
upper = np.array(list(string.ascii_uppercase))
lower = np.array(list(string.ascii_lowercase))
digits = np.array(list(string.digits))
special = np.array(list(string.punctuation))

# Combine all character sets
all_chars = np.concatenate((upper, lower, digits, special))

n = int(input("Password length: "))
m = int(input("Number of passwords: "))

if n < 4:
    print("Password length must be at least 4.")
else:
    passwords = []

    for _ in range(m):
        password = []

        # Guarantee one character from each category
        password.append(np.random.choice(upper))
        password.append(np.random.choice(lower))
        password.append(np.random.choice(digits))
        password.append(np.random.choice(special))

        # Fill the remaining characters from the combined pool
        remaining = n - 4
        if remaining > 0:
            password.extend(np.random.choice(all_chars, size=remaining))

        # Shuffle using NumPy
        password = np.array(password)
        np.random.shuffle(password)

        passwords.append("".join(password))

    print("\nGenerated Passwords:\n")
    for i, pwd in enumerate(passwords, start=1):
        print(f"{i}. {pwd}")
```

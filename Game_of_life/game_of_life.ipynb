

def count_neighbors(grid, row, col):
    rows, cols = grid.shape
    count = 0

    for dr in [-1, 0, 1]:
        for dc in [-1, 0, 1]:

            if dr == 0 and dc == 0:
                continue

            nr = row + dr
            nc = col + dc

            if 0 <= nr < rows and 0 <= nc < cols:
                count += grid[nr][nc]

    return count


def next_generation(grid):
    rows, cols = grid.shape
    new_grid = np.zeros((rows, cols), dtype=int)

    for row in range(rows):
        for col in range(cols):

            neighbors = count_neighbors(grid, row, col)

            if grid[row][col] == 1:

                if neighbors == 2 or neighbors == 3:
                    new_grid[row][col] = 1

            else:

                if neighbors == 3:
                    new_grid[row][col] = 1

    return new_grid


def simulate(grid, generations=15, delay=0.3):
    plt.figure(figsize=(6,6))

    for generation in range(generations):
        plt.clf()
        plt.imshow(grid, cmap="binary")
        plt.title(f"Generation {generation + 1}")
        plt.axis("off")
        plt.pause(delay)

        grid = next_generation(grid)

    plt.show()

```python
import numpy as np
import matplotlib.pyplot as plt

def next_generation(grid):

    padded = np.pad(grid, 1)

    neighbors = (
        padded[:-2, :-2] +
        padded[:-2, 1:-1] +
        padded[:-2, 2:] +
        padded[1:-1, :-2] +
        padded[1:-1, 2:] +
        padded[2:, :-2] +
        padded[2:, 1:-1] +
        padded[2:, 2:]
    )

    survive = (grid == 1) & ((neighbors == 2) | (neighbors == 3))
    born = (grid == 0) & (neighbors == 3)

    return (survive | born).astype(int)


def simulate(grid, delay=0.1):

    fig, ax = plt.subplots(figsize=(8, 8))

    img = ax.imshow(grid, cmap="binary", interpolation="nearest")

    ax.set_title("Conway's Game of Life")
    ax.axis("off")
    generation=0

    while True:
        img.set_data(grid)
        ax.set_title(f"Generations {generation}")
        fig.canvas.draw_idle()

        

        
        plt.pause(delay)
        grid = next_generation(grid)

        generation+=1

        

grid = np.random.choice([0, 1], size=(50, 50), p=[0.7, 0.3])

simulate(grid)
```
